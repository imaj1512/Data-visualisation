# Data Visualization


**Bar plot**:

Bar plot is the plot which reresent categorical data

```python
df['diabetes'].value_counts().plot.bar(title='less chance of diabetes vs more chance of diabetes')
plt.xlabel('diabetes')
plt.ylabel('count')
plt.legend()
```

**Observation**:


*   No of people with more chance of diabetes is comparitvely less




**Hist** **plot**

```python
age=df['age'].tolist()

plt.xlabel('ages',color='m')
plt.ylabel('counts',color='m')
plt.title('ages distribution',color='k',fontdict={'fontsize':10,'family':'fantasy','color':'darkred','weight':'normal'})
plt.style.use('classic')

plt.hist(age,bins=[20,30,40,50,60,70,80],color='g',label='age',rwidth=0.5)

plt.legend(loc=0,fancybox=True)
plt.show()
```

**Observation**

* In between age 50 to 60 peoples more in our data which is more than 14000


```python
gen=sns.countplot(x='diabetes',data=df,hue='gender')

for bars in gen.containers:
    gen.bar_label(bars)
```

**Observation:**


Diabetes can be diagnosed irrespective of gender

**Pie chart**



Percentage of Non_diabetic vs Diabetic patients



```python

df['diabetes'].value_counts().plot(kind='pie',autopct='%0.2f')
plt.title("Distribution of diabeties in dataset")
```

**Observation:**

Most of them are non-diabetic

**Heat map**

```python
fig,x=plt.subplots(figsize=(10,8))
sns.heatmap(df.corr(),annot=True,linewidth=0.5,ax=x)
plt.plot()
```

**Observation:**

The gender and smoke history least correlated.

The highest correlation observed between glucose level and diabetes.

**Box plot**

Highlighting The major factors that causes diabetes using boxplot

```python
sns.boxplot(x="diabetes", y="bmi", data=df)
plt.title("BMI Distribution by Diabetes Status")
plt.show()

sns.boxplot(x="diabetes", y="HbA1c_level", data=df)
plt.title("HbA1c Level Distribution by Diabetes Status")
plt.show()

sns.boxplot(x="diabetes", y="blood_glucose_level", data=df)
plt.title("Blood Glucose Level Distribution by Diabetes Status")
plt.show()
```

**Observation:**
The above Boxplots gives us the idea that people having bmi over 30 ,and hbac level over 6 and blood glucose level over 150 are having diabetes

**Violin plot**

```python
sns.violinplot(x="hypertension",y="age",data=df,hue='diabetes')
```
